#### Task 1
Find out top-20 of most popular artist's in terms of: raw listens, unique users listens; 

##### At first two datasets must be merged by 'join' function. As we are interested in 'artist_name', the dataset should be grouped by this feature. Then to find 'raw listens' number of users who listen each song are be summed up. In case to find 'unique users listens' only unique users must be counted. 
##### Finally the dataset ordered by two columns 'raw listens' and 'unique users listens' to show top 20 artist in terms of all listens and unique ones.

```python
joined = listen_history.join(songs, ["song_id"], 'inner')

joined.groupBy("artist_name")\
.agg(F.count(F.col("msno")).alias("raw listens"), F.countDistinct(F.col("msno")).alias("unique users listens"))\
.orderBy(F.col("raw listens").desc(), F.col("unique users listens").desc())\
.limit(100).toPandas().head(20)
```

#### Task 2
Find out top-3 languages by num of raw listens

##### Basically the same logic applied in this task, since we need to join datasets, then aggregate it by number of all listens and sort by it to find top 3 artist in terms of raw listens.

```python
joined = listen_history.join(songs, ["song_id"], 'inner')

joined.groupBy("language")\
.agg(F.count(F.col("msno")).alias("raw listens"))\
.orderBy(F.col("raw listens").desc())\
.limit(100).toPandas().head(3)
```

#### Task 
##### Build histogram of songs length by language

```python
# create a dataframe where language aggregated by sum of songs lengths
language_song_length = songs.select("language", "song_length")\
.groupby(F.col("language"))\
.agg(F.sum("song_length").alias("song_length_sum"))\
.toPandas()
```

```python
# take a look at the dataframe
language_song_length.head(5)
```

```python
# add column with song length in minutes, because now it's in milliseconds
language_song_length["song_length_sum_minute"] = language_song_length["song_length_sum"]/60000
language_song_length.sample(10)
```

```python
# check if there are missing values
language_song_length.isna().sum()
```

```python
# one missing value is not crutial in this case, but still let's delete it
language_song_length = language_song_length[~language_song_length['language'].isna()]
```

```python
# sort dataframe by song length
language_song_length = language_song_length.sort_values(by='song_length_sum', ascending=False)
language_song_length.head(5)
```

```python
# building a barplot 
fig, ax = plt.subplots(figsize=(16, 6))

# depicting the visualization
sns.barplot(x='language', y='song_length_sum_minute', data = language_song_length, ax=ax, palette="Set3")

plt.title('Language by sum of song lengths in minutes')
plt.ylabel('Sum of song lengths in minutes')

plt.show()
```
